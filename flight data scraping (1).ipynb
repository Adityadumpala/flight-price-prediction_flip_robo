# data of date 29/05/2022, economy class, delhi to mumbai

```python
#flight name
name_flight=[]
name_flght=driver.find_elements_by_xpath('//span[@class="txt-r4 ng-binding"]')

for i in name_flght:
    name_flight.append(i.text)
    
#source and destination
source_destination=[]
src_dstn=driver.find_elements_by_xpath('//span[@class="txt-r3-n ng-binding"]')
for i in src_dstn:
    source_destination.append(i.text)

source=source_destination[::2]
destination=source_destination[1::2]

#no of stops
stops=[]
no_of_stops=driver.find_elements_by_xpath('//span[@class="dura_md2 ng-scope"]')
for i in no_of_stops:
    stops.append(i.text)
    

#travelling time
duration=[]
drtion=driver.find_elements_by_xpath('//span[@class="dura_md ng-binding"]')
for i in drtion:
    duration.append(i.text)

#price
price=[]
prc=driver.find_elements_by_xpath('//div[@class="col-md-8 col-sm-8 col-xs-9 txt-r6-n exPrc ng-binding"]')

for i in prc:
    price.append(i.text)
    
#starting and ending times
time=[]
timings=driver.find_elements_by_xpath('//span[@class="txt-r2-n ng-binding"]')

for i in timings:
    time.append(i.text)

start_time=time[::2]
end_time=time[1::2]
    
print(len(name_flight),len(source_destination),len(stops),len(duration),len(price),len(time),len(source),len(destination),len(start_time),len(end_time))
```

```python
import pandas as pd
data1=pd.DataFrame({'flight_name':name_flight,'source':source,'destination':destination,'no_of_stops':stops,'start_time':start_time,'end_time':end_time,'travel_time':duration,'price':price})
data1['date_of_journey']='29/5/2022'
data1['class']='Economy'
data1.head()
```

```python
data1.shape
```
